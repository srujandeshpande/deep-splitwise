```python
import requests
from dotenv import load_dotenv
import matplotlib

load_dotenv()

import os
key = os.getenv("API_KEY")
```

```python
url = "https://secure.splitwise.com/api/v3.0/get_current_user"

payload={}
headers = {
  'Authorization': 'Bearer '+key
}

response = requests.request("GET", url, headers=headers, data=payload)

print(response.text)
```
