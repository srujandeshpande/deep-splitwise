```python
import pandas as pd 
import matplotlib.pyplot as plt
from itertools import islice
```

```python
df = pd.read_csv("csv/7-am-at-spice-garden_2021-03-25_export.csv") 
# print(df)
```

```python
df2 = df.loc[df['Category'] == 'Gas/fuel']

# print(df2)

# df2['count'] = df.select_dtypes(include='float64').gt(0).sum(axis=1)
# counts = pd.DataFrame(df2['full_name'].value_counts(), columns=['num_unique'])
names = []
days = []

for i in df2.columns[5:]:
    # print(i)

    df22 = df2.apply(lambda x : True
                if x[i] > 0 else False, axis = 1)
    num_rows = len(df22[df22 == True].index)
    names.append(i)
    days.append(num_rows)

print(names, days)
```

```python
fig = plt.figure()
plt.barh(names,days)
plt.xticks(rotation=90)
for index, value in enumerate(days):
    plt.text(value+0.1, index, str(value))
plt.title("Days driven to college")

plt.show()
```
